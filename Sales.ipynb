```python
sales2=pd.read_csv(r'C:\Users\abhij\Desktop\pycon-pandas-tutorial-master\sales2.csv')
sales2.fillna('')
```

```python
sales2.fillna('')
```

```python
df1=sales1.rename(columns={'Book title':'title','Number sold':'number','Sales price':'their_price','Royalty paid':'we_got'})
df1['currency']='USD'
df1
```

```python
s = sales2.copy()
t = sales2['Title']
t = t.where(t.str.endswith(')')).str.split().str[-1].str.strip('()')
s['currency'] = t.fillna(method='bfill')
s = s[s['List price'].notnull()]
df2 = s.rename(columns={
    'Units sold': 'number',
    'Title': 'title',
    'List price': 'their_price',
    'Royalty': 'we_got',
    })
s.head()
```

```python
df = pd.concat([df1, df2])
df
```

```python
df['total_paid'] = df.their_price * df.number
df.groupby(['title', 'currency'])[['total_paid']].sum()
```
